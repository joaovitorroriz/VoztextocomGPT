# Transformers

Transformers foram introduzidos com o paper "*Attention is All You Need*" em 2017 (https://proceedings.neurips.cc/paper_files/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf) por pesquisadores da Google.

Eles foram concebidos, inicialmente, para a tradução de texto de uma língua para outra.

Na época, os modelos dominantes de transdução de sequências de texto (ou outros tipos de sequências) baseavam-se em redes neurais recorrentes ou convolucionais complexas, que incluiam um codificador e um decodificador.

Os modelos de melhor desempenho também conectavam o codificador e o decodificador por meio de um mecanismo de atenção. No paper "*Attention is All You Need*" uma nova arquitetura de rede simples, o Transformer, baseada unicamente em mecanismos de atenção, dispensando completamente a recorrência e as convoluções foi apresentada.

Experimentos em duas tarefas de tradução automática mostraram que transformers são superiores em qualidade, sendo mais paralelizáveis e exigindo significativamente menos tempo para treinar.

O tranformer alcançou desempenho superior na tarefa de tradução do Inglês para o Alemão com apenas uma pequena fração dos custos de treinamento dos melhores modelos da literatura na época.

# Generative Pre-trained Transformers (GPT)

Generative Pre-trained Transformers (GPT) compreendem uma série de modelos de linguagem baseados na arquitetura Transformer. No entanto, ao contrário do Transformer original, GPTs usam apenas a parte do decodificador da arquitetura.
   - **Treinamento em Duas Fases**: O GPT é treinado em duas fases:
     1. **Pré-treinamento**: O modelo é treinado como um modelo de linguagem em um grande corpus para prever a próxima palavra em uma sequência.
     2. **Ajuste Fino**: O modelo pré-treinado é então ajustado para tarefas específicas, como classificação de texto, resposta a perguntas, etc.
   - **Aplicação**: GPT e suas variantes (como GPT-2, GPT-3 e ChatGPT) são conhecidos por serem modelos de linguagem altamente eficazes e têm sido aplicados em uma ampla variedade de tarefas de processamento de linguagem natural.

**Em resumo**: "Transformers" refere-se à arquitetura global que introduziu mecanismos de atenção e auto-regressão no processamento de linguagem natural. "Generative Pre-trained Transformers (GPT)" refere-se a uma aplicação específica e uma abordagem de treinamento de modelos baseados na arquitetura Transformer.

# Atenção e Auto-Regressão em GPTs


1. **Mecanismo de Atenção**:
    - Em um Transformer, a atenção refere-se à capacidade do modelo de ponderar diferentes partes de uma sequência de entrada ao gerar uma saída. O mecanismo de atenção permite que o modelo decida a quais partes da entrada dar mais "atenção" com base no contexto.
    - O mais notável é o mecanismo de "atenção auto-atentiva" ou "self-attention", que permite que cada item em uma sequência considere outros itens da mesma sequência para determinar sua representação.

2. **Modelos Auto-regressivos**:
    - Modelos auto-regressivos são modelos que geram saídas sequencialmente. Quando geram uma sequência, a saída em um determinado passo de tempo é condicionada nas saídas geradas nos passos anteriores.
    - GPT é um exemplo de modelo auto-regressivo. Ao gerar texto, ele produz uma palavra de cada vez, e a próxima palavra é condicionada nas palavras anteriores geradas pelo modelo.


- O mecanismo de **atenção** no Transformer permite que o modelo atribua pesos variáveis a diferentes partes da entrada.
- **Auto-regressivo** descreve uma abordagem de geração sequencial onde cada passo depende dos passos anteriores.

A linguagem é a base da comunicação humana e desempenha um papel vital na formação de nossas interações com o mundo.

Com o advento do PLN (Processamento de Linguagem Natural), revolucionou-se a forma como interagimos com máquinas. O PLN tornou-se um divisor de águas no mundo da comunicação, permitindo que humanos interajam com máquinas de uma forma mais natural.

A evolução do PLN foi alimentada pelo crescimento exponencial de dados textuais na internet. Ao longo dos anos, o PLN testemunhou uma significativa transformação, de sistemas simples baseados em regras para modelos complexos baseados em aprendizado profundo.

Apesar dos avanços, a compreensão e geração de linguagem natural vinha sendo um problema desafiador no campo do PLN, em grande parte devido à natureza complexa da linguagem humana.

No entanto, avanços recentes abriram caminho para novas abordagens para enfrentar esses desafios. Uma dessas inovações no PLN é o desenvolvimento de GPTs.

GPTs tornaram-se famosas após o lançamento do ChatGPT pela OpenAI, uma empresa de pesquisa focada no desenvolvimento de tecnologias de IA.

O GPT é um modelo de aprendizado profundo que é pré-treinado em grandes corpora de dados textuais e pode ser ajustado para tarefas específicas como geração de linguagem, análise de sentimentos, modelagem de linguagem, tradução automática e classificação de texto.

A arquitetura transformer usada em GPTs é um avanço significativo em relação às abordagens anteriores para PLN, como RNN e CNN. Ela utiliza um mecanismo de autoatenção para permitir que o modelo considere o contexto da frase inteira ao gerar a próxima palavra, o que melhorou a capacidade do modelo de entender e gerar linguagem.

GPTs podem desempenhar uma ampla gama de tarefas em PLN. Uma de suas principais forças está na compreensão de linguagem natural (NLU), onde pode analisar e compreender o significado do texto, incluindo a identificação de entidades e relações nas frases.

São também proficientes na geração de linguagem natural (NLG), o que significa que podem criar saídas de texto, como escrever conteúdo criativo ou responder perguntas de forma abrangente e informativa.

GPTs também são um geradoras de código, sendo capazes de escrever código em várias linguagens, como Python ou JavaScript.

GPTs também pode ser utilizadas para resposta a perguntas, o que significa que podem fornecer resumos de tópicos factuais, criar histórias com base no texto de entrada ou resolver problemas apresentados textualmente.

Além disso, GPTs podem resumir um texto, fornecendo uma visão geral de um artigo de notícias ou trabalho de pesquisa, e podem ser usadas para tradução, o que torna possível traduzir texto de um idioma para outro.

Em geral, a capacidade de GPTs de realizar uma ampla gama de tarefas de PLN com alta precisão as torna uma ferramenta inestimável para várias indústrias, incluindo finanças, saúde, marketing e mais.

À medida que a tecnologia PLN continua avançando, podemos esperar que o GPT e outros modelos de linguagem se tornem ainda mais sofisticados e poderosos, permitindo-nos comunicar com máquinas de forma mais natural e eficaz.

Isso tudo sem mencionar sua capacidade de compreender e manipular imagens, sons, entre outras modalidades de dados.
